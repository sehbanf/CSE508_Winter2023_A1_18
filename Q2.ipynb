```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):

  #lowercase the text
  text = text.lower()

  #remove punctuation
  text = re.sub(r'[+*/\\\-?.>,<\"\';:!@#$%^&()_`~]', ' ', text)
  
  #tokenize
  tokens = word_tokenize(text) #tokens will contain list of tokens

  #remove stopwords
  tokens = [token for token in tokens if token not in stop_words]

  tokens = [token for token in tokens if token!=" "]

  return tokens
```

```python
def OR_opr(x,y):
  #since x and y is sorted hence we can merge in O(x+y) time

  res = []

  comp = 0

  x_ptr,y_ptr = 0,0

  n = len(x)
  m = len(y)

  while(x_ptr<n and y_ptr<m):

    if x[x_ptr] == y[y_ptr]:
      res.append(x[x_ptr])
      x_ptr = x_ptr + 1
      y_ptr = y_ptr + 1
      comp = comp + 1
    
    else:
      comp = comp + 2 # first comp for equality and then for the ineq

      if x[x_ptr]>y[y_ptr]:
        res.append(y[y_ptr])
        y_ptr = y_ptr + 1
      
      else:
        res.append(x[x_ptr])
        x_ptr = x_ptr + 1
  

  while(x_ptr<n):
    res.append(x[x_ptr])
    x_ptr = x_ptr + 1
  
  while(y_ptr<m):
    res.append(y[y_ptr])
    y_ptr = y_ptr + 1
  

  return res,comp
  

```

```python
def AND_opr(x,y):

  #note: both x and y are sorted
  n = len(x)
  m = len(y)

  x_ptr,y_ptr=0,0

  res = []
  comp = 0

  while(x_ptr<n and y_ptr<m):

    if x[x_ptr] == y[y_ptr]:
      res.append(x[x_ptr])
      
      x_ptr = x_ptr + 1
      y_ptr = y_ptr + 1

      comp = comp + 1
    
    else:
      comp = comp + 2

      if x[x_ptr]<y[y_ptr]:
        x_ptr = x_ptr + 1
      
      else:
        y_ptr = y_ptr + 1
  

  return res,comp
        

```

```python
#Constructing the universe

list_of_doc_ids = list(inverted_index.values())

U = set()

for list_of_id in list_of_doc_ids:
  U.update(list_of_id)
```

```python
def MINUS(q):
  
  #subtracts q from all the doc ids in the corpus
  return list(U.difference(set(q)))
```

```python
def OR_NOT_opr(x,y):
  
  not_y = MINUS(y)
  not_y = sorted(not_y)

  res,comp = OR_opr(x,not_y)

  return res,comp
```

```python
def AND_NOT_opr(x,y):
  
  not_y = MINUS(y)
  not_y = sorted(not_y)

  res,comp = AND_opr(x,not_y)

  return res,comp
```

```python
def get_inp():

  print("Execute Query ------>\n")

  N = int(input("Enter the number of queries to execute -->"))

  list_of_ip_seq = []
  list_of_op_seq = []

  for i in range(N):
    input_seq = input("\nEnter the input sequence ---->")
    op_seq = input("\nEnter the operation sequence ---->")

    list_of_ip_seq.append(input_seq)
    list_of_op_seq.append(op_seq)
  
  return list_of_ip_seq,list_of_op_seq




```

```python
def check(list_of_tok):

  for tok in list_of_tok:
    if tok not in inverted_index.keys():
      return False

  return True 
```

```python
def main():

  list_of_ip_seq,list_of_op_seq = get_inp()

  for index, (ip_seq, op_seq) in enumerate(zip(list_of_ip_seq,list_of_op_seq)):
    
    list_of_tokens = preprocess(ip_seq)

    #check if the tokens in query is present in our corpus    
    if not check(list_of_tokens):
      print("The Query cannot be computed as the tokens in the query do not exist in the corpus")
      continue #hence moving onto the next query
      


    list_of_op = op_seq.split(",")
    
    if len(list_of_tokens)-1 != len(list_of_op):

      print("Query cannot be computed , number of tokens not matching with the operators")
      print("\nList of tokens", list_of_tokens)
      print("\nlist_of_op",list_of_op)
      continue
    
    ## Constructing the query string
    q_str = list_of_tokens[0] + " "

    for tok,op in zip(list_of_tokens[1:],list_of_op):
      q_str = q_str + op + " " + tok + " "
    
    q_str = q_str.strip()

    print(f" \nQuery {index + 1} : {q_str}")


    ################## COMPUTING THE QUERY #########################
    res = list_of_tokens[0]
    res_doc_ids = inverted_index.get(res)
    
    t_c = 0
    
    for tok,op in zip(list_of_tokens[1:],list_of_op):
      tok_doc_ids = inverted_index.get(tok)
    
      if op == "OR":
        res_doc_ids,comp = OR_opr(res_doc_ids,tok_doc_ids)
      
      if op == "AND":
        res_doc_ids,comp = AND_opr(res_doc_ids,tok_doc_ids)
      
      if op == "AND NOT":
        res_doc_ids,comp = AND_NOT_opr(res_doc_ids,tok_doc_ids)
      
      if op == "OR NOT":
        res_doc_ids,comp = OR_NOT_opr(res_doc_ids,tok_doc_ids)
      
      t_c = t_c + comp

    ##Finding the number of docs
    print(f"\nNumber of documents retrieved for query {index + 1} ----> ",len(res_doc_ids))

    #Print the file names of the docs
    file_names = []
    for doc_id in res_doc_ids:
      file_names.append(doc_map[doc_id])

    print(f"\nNames of the documents retrieved for query {index + 1} ----> ",file_names)

    #Print the number of comparisons required for the query
    print(f"\nNumber of comparisons required for query {index + 1} ------> ",comp)


```

```python
main()
```
