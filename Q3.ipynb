# (i) Bigram inverted index [30 Marks]

1. Create a bigram inverted index (from scratch; No library allowed) of the dataset obtained
from Q1.


2. Use Python’s pickle module to save and load the bigram inverted index.

```python
doc_map = dict()


for index,doc in enumerate(os.listdir(PATH_TO_PREPROCESSED_DATA)):
  doc_map[index] = doc
```

```python
len(os.listdir(PATH_TO_PREPROCESSED_DATA))
```

make a list of all bigrams in the corpus

```python
bigrams = set()

for doc in os.listdir(PATH_TO_PREPROCESSED_DATA):
  
  with open(os.path.join(PATH_TO_PREPROCESSED_DATA,doc),'r') as f:
    data = f.read()
    data = data.split() # list of unigrams
    
    bigrams_doc = [] #list of tupples (word1,word2)
    
    for i in range(len(data)-1):
      bigram = data[i],data[i+1]
      bigrams_doc.append(bigram)
    
    bigrams.update(bigrams_doc)

    
    

  

```

```python
len(bigrams)
```

###### Construction of the inverted bigram index

```python
file_ids = sorted(doc_map.keys())
```

```python
def read_file(file_path):
  
  with open(file_path,'r') as f:
    text = f.read()
  
  return text
```

```python
inverted_bigram_index = dict()

for bigram in tqdm(bigrams):
  
  for id in file_ids:
    
    #read the text from that file
    file_path = doc_map[id]
    file_path = os.path.join(PATH_TO_PREPROCESSED_DATA,file_path)

    text = read_file(file_path)

    bigram_text = bigram[0]+ " " + bigram[1]

    if bigram_text in text:
      
      if bigram in inverted_bigram_index:
        inverted_bigram_index[bigram].append(id)
      
      else:
        inverted_bigram_index[bigram] = [id]
    

print(inverted_bigram_index)




```

```python
PATH_TO_STORE_INVERTED_BIGRAM_INDEX = "./Saved/inverted_bigram_index.pickle"
```

```python
inverted_bigram_index 
```

```python
with open(PATH_TO_STORE_INVERTED_BIGRAM_INDEX,'wb') as f:
    pickle.dump(inverted_bigram_index,f,protocol=pickle.HIGHEST_PROTOCOL)
```
