## (ii) Positional index

1. Create a positional index (from scratch; No library allowed) of the dataset obtained from
Q1.
2. Use Python’s pickle module to save and load the positional index.

make a list of all tokens in the corpus

```python
tokens = set()

for doc in os.listdir(PATH_TO_PREPROCESSED_DATA):
  
  with open(os.path.join(PATH_TO_PREPROCESSED_DATA,doc),'r') as f:
    data = f.read()
    data = data.split() # list of unigrams
    
    tokens.update(data)
```

```python
len(tokens)
```

```python
file_ids = sorted(doc_map.keys())
```

```python
def read_file(file_path):
  
  with open(file_path,'r') as f:
    text = f.read()
  
  return text
```

```python
def find_pos_freq(text,token):
    
    re_object = re.finditer(pattern=token, string=text)
    indices = [index.start() for index in re_object] # will return a list of pos
    
    return len(indices),indices
    
```

```python
pos_index = dict()

## form {token:[freq_of_token,{doc_id:[pos1,pos2,pos3],doc_id2:[pos1,pos2,pos3]}]}

for token in tqdm(tokens):
    
    t_freq = 0
    
    for f_id in file_ids:
    
        #read the text from that file
        file_path = doc_map[f_id]
        file_path = os.path.join(PATH_TO_PREPROCESSED_DATA,file_path)

        text = read_file(file_path)

        
        if token in text:
            
            freq,pos = find_pos_freq(text,token) # freq = freq of token in text and list of pos of token in text
        
            t_freq = t_freq + freq

            if token in pos_index:
                pos_index[token][0] = t_freq
                pos_index[token][1].append({f_id:pos})
                
            
            else:
                pos_index[token] = [t_freq,[{f_id:pos}]]
    
    




```

```python
print(pos_index)
```

Save the pos index

```python
PATH_TO_SAVED = "./Saved/pos_index.pickle"
```

```python
with open(PATH_TO_SAVED,'wb') as f:
    pickle.dump(pos_index,f,protocol=pickle.HIGHEST_PROTOCOL)
```
